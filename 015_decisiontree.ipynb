```python
#랜덤포레스트 - bagging의 앙상블 학습이다.
#bagging = bootstrap + aggregation
#디시젼트리의 단점을 극복하기 위해서( 변수의 순서가 바뀌면 결과도 달라진다. 과적합 우려된다.)
#평균과 투표로 결과가 도출된다.
#모델투표로 바뀐다. model voting 시스템
#RF - tree모델을 모아 놓은것 
#model voting - tree, gbm, ada boost, ann 등 이종의 모델을 투표하게 되는것(각모델이 학습을 하고 그놈을 모아서 투표를 하게된다.)
#항상 홀수 모델 갯수를 사용하게 된다. (2:2같은 투표 결과가 나오면 안되기 때문에)
#범주화 0또는 1로 결과를 분류해서 다수의 투표를 받는 결과를 택한다.
#voting/stacking -> 투표/이종 모델들의 결과를 다시학습하는 final model을 갖고있다.

from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
#model_selection - spliter, hyper parameter optimizers, model validatrion
from sklearn.feature_selection import SelectFromModel #변수중요도에 따라 모델을 선택한다.
#특징선택
from sklearn.metrics import accuracy_score #recall_score - 민감도, precison_score - 정밀도, specificity_score - 특이도, 
#f1_score(중요!) 민감도와 고려한 것으로 중요하다. 많이 사용된다.
iris = datasets.load_iris()
fear_labels = ['Sepal Length','Sepal Width','Petal Length','Petal Width']

x = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.4, random_state=0)
clf = RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1) #10000개의 예측기 = 모델을 10000개 만들어라(랜덤으로)
#bootstrap을 랜덤으로 한다.(특징을 조합)
#n_jobs=-1 분산처리 연산기의 core의 갯수를 다써라.(cpu의 코어를 다써라)
clf.fit(X_train, y_train)
for feature in zip(fear_labels, clf.feature_importances_):
    print(feature)
    #두개를 묶어서 출력
sfm = SelectFromModel(clf, threshold=0.15) #변수중요도를 이용해서 모델을 선택
#threshold 문지방 - 이 기준을 넘어서는 놈을 선택한다.
sfm.fit(X_train, y_train)
for feature_list_index in sfm.get_support(indices=True):
    #threshold를 통과한 특성 - sfm.get_support(indices=True) 변수명으로 주지말고 인덱스로 줘라
    print(fear_labels[feature_list_index])
X_importance_train = sfm.transform(X_train) #변수를 선택하라(특징추출, 문지방을 넘어선 놈으로 선택하라.)
#scikits에서는 일관된 인터페이스를 가지고 있다.
#어떤모델이든지 fit->transform을 하게된다. or fit_transform
X_importance_test = sfm.transform(X_test) #test에도 해준다.
clf_importance = RandomForestClassifier(n_estimators=1000, random_state=0, n_jobs=-1)
clf_importance.fit(X_importance_train, y_train) #fitting하게 된다.
#전체변수가 다 있는 모델
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
#중요한 변수가 선택된 모델
y_importance_pred = clf_importance.predict(X_importance_test)
accuracy_score(y_test, y_importance_pred)
```

```python
#imputer & cross_val_score
#imputer->impute 변경

from sklearn.datasets import load_boston
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline #프로세스의 터널을 만드는것(묶음)
from sklearn import impute #결측치 보충, 대체
from sklearn.model_selection import cross_val_score
#결측치 처리방식 - 어느것이 나은지 보는것이다.

rng = np.random.RandomState(0)
dataset = load_boston() #bunch 자료형 (scikits 전용 포맷)
X_full, y_full = dataset.data, dataset.target
n_samples = X_full.shape[0] #행수(데이터 수)
n_features = X_full.shape[1] #열수(변수)

estimator = RandomForestRegressor(random_state=0, n_estimators=100) #100개의 트리를 생성한다.(DT regressor을 100개 묶음)
score = cross_val_score(estimator, X_full, y_full).mean() #cross_val_score 교차검증을 하게된다. (디폴트 5개 만듬)
#5번으로 자른다음 진행해서 각기다른 스코어를 평균을 내라
print('전체데이터에 대한 평가 = %.2f'%score)

missing_rate = 0.75
n_missing_samples = int(np.floor(n_samples * missing_rate)) #총데이터수*0.75 -> 결측치로 만들 갯수
#h - horizontal, vertical 
missing_samples = np.hstack((np.zeros(n_samples - n_missing_samples, dtype=np.bool), np.ones(n_missing_samples, dtype=np.bool)))
#전체 데이터에 대해서 75퍼센트는 1을주고 나머지는 0을줌 이진으로 만들었음(t/f 조합을 만듬) -> 불린인덱싱 하려고
rng.shuffle(missing_samples) #0과1을 섞어줌
missing_features = rng.randint(0, n_features, n_missing_samples) #피쳐수에서 미싱샘플수만큼 만들어라
X_filtered = X_full[~missing_samples, :] #실제데이터만 남게됨 (결측치를 뺌) '~'는 참과거짓을 반대로
y_filtered = y_full[~missing_samples] #같으나 x는 열도 여러개니까 여기는 인덱스 하나만 사용
estimator = RandomForestRegressor(random_state=0, n_estimators=100)
score = cross_val_score(estimator, X_filtered, y_filtered).mean()
print('결측치 제거후 = %.2f'%score) #결측치를 제거하였을때 - 행 하나가 삭제된다. 

X_missing = X_full.copy() #메모리 복사
X_missing[np.where(missing_samples)[0], missing_features]=0 #미싱위치에 0을대입해라
y_missing = y_full.copy()
estimator = Pipeline([('imputer', impute.SimpleImputer(missing_values=0, strategy='mean')), 
                      #랜덤포레스트 하기전에 미싱벨류가 0인놈을 평균으로 채워라
                      #이 모델을 이름을 부여하여 참조할 수 있다.(파이프 라인 안에서)
                      #grid search cv - 파라메터 튜닝을 한다.(파라메터 조합 생성)
                      #이름을 부여해서 안의 매개변수를 변화하여 쓰기 위해서 이름을 준 것이다.
                      ('forest',RandomForestRegressor(random_state=0, n_estimators=100))])
#100개 만들어라
#파이프 라인도 모델이다.
score = cross_val_score(estimator, X_missing, y_missing).mean()
#cross_val_score 매개변수는 (모델 명, 훈련데이터, 타깃, cv)인데, 여기에서 cv는 폴드(fold)수
print('결측치를 평균값으로 대체한후 처리 score =%.2f'%score)

#결측치에 값을 채워주는 놈이 임퓨터
#전체데이터와 평균값으로 대체된것은 결측치의 갯수가 달라져도 그게 영향받지 않지만 제거하는경우는 많이 영향받는다.
#모델의 안정성이 있어야한다.
```

```python
#모델, 독립변수(2개), 종속변수, ax=도화지(디폴트매개변수), cmap(디폴트매개변수)
#독립 변수갯수를 2개가 넘으면 사용불가능 일반화가 안된다.
#시각화 - matplotlib, seaborn, manifod(다형체,다양체)-MDS(멀티디멘션 스케일, 다차원 척도법)
#MDS선형적이라 MRS(move,rotate,scale)에 강하다.
#LLE - 주병의 k개의 근접한 거리값만 고려
#비선형 - isomap
#다차원 군집분석 - TSNE
#위의놈들은 transform이 안된다. 축소해놓고 원래 차원으로 돌아가려면 불가능, 시각화 용으로 쓴다.

#decomposition : PCA - inverse_transform이 통한다.(차원축소후 원래대로 돌아갈 수 있다., 분석에서 사용가능)
    #FA(facort analysis - 비슷한 변수는 묶어서 하나로 표현한다.)
    #비정방 매트릭스를 디컴퍼지션하는놈 SVD(singular value decomposition)

#정방행렬 분해 - eigen분해(고유값분해)
#SVD 비정방 행렬의 의사해를 구해준다.(요긴함) 해석학을 학습할 때 많이 쓰인다.

def visual_classifier(model, X, y, ax=None, cmap='rainbow'):
    ax = ax or plt.gca() #매개변수로 안들어오면 plt에서 얻어와라
    ax.scatter(X[:,0],X[:,1],c=y,s=30,cmap=cmap)
    ax.axis('tight') 
    ax.axis('off') #축은 표현하지 말고(이미지로 표현하고 싶을때 보통 사용하지않음)
    xlim = ax.get_xlim() #범위
    ylim = ax.get_ylim()
    model.fit(X,y) #들어온 모델로 학습
    xx,yy = np.meshgrid(np.linspace(*xlim, num=200), np.linspace(*ylim, num=200))
    #200*200=40000개 조합으로 출력된다.(사각형)
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    #ravel = flatten 과 같음(차수를 없앤다, 메모리에 있는 1차원 그래도 가지고 오겠다.)
    #predict는 차수를 없애고 넣어줘야 한다.
    #점하나마다 값을 예측했다.
    n_classes = len(np.unique(y)) #종류가 몇개냐? 종속변수의 갯수가 몇개냐(범주만 출력된다. 반복제거)
    contours = ax.contourf(xx,yy,Z,alpha=0.3,levels=np.arange(n_classes+1)-0.5, cmap=cmap,clim=(y.min(),y.max()), zorder=1)
    #contourf 등고선+fill 채워서 등고선 그려라
    #zorder=1 레이어의 순서, 이미지에서의 레이어
    ax.set(xlim=xlim, ylim=ylim)
```

```python
from sklearn.model_selection import GridSearchCV
#파라메터 튜닝(기능) - 최적점을 찾기 위해서(목적)
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import make_scorer #매트릭스에서 어떤점수로 할 것인지 점수 함수를 생성
from sklearn.preprocessing import MinMaxScaler 
#standard scaler
#minmax scaler
#robust scaler
#nomalizer scaler
from sklearn.metrics import roc_auc_score #true positive rate/ false positive rate 이루는 모든 점에서의 조합도
#auc - 면적계산

cancer = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(cancer.data, cancer.target, random_state=0)
scaler = MinMaxScaler().fit(X_train) #최고값 최소값을 갖고있다가.
#정규화 , 종속변수 스케일링
X_train_scaled = scaler.transform(X_train)
learner = RandomForestClassifier(random_state=2) #RFC 분류한다.(목적)
n_estimators=[12,24,36,48,60]
min_samples_leaf=[1,2,3,8,16] #5*5 = 25개의 조합이 만들어졌다.
parameters = {'n_estimators':n_estimators, 'min_samples_leaf':min_samples_leaf}
#두배열을 묶어 dict형식으로 생성
#예측수 몇개로 해야하는지, 말단의 샘플수를 몇개로 할건지 (어떤놈이 좋은지 배열로 돌려봄)

def auc_scorer(target_score, prediction):
    #개인이 원하는대로 customizing
    auc_value = roc_auc_score(prediction, target_score)
    return auc_value

scorer = make_scorer(auc_scorer, greater_is_better=True) #점수를 커스터마이징 한 경우 make_scorer로 감싸준다(wrapper)
grid_obj = GridSearchCV(learner, parameters, scorer) #cv할때 척도를 무엇으로 할 것이냐(어떤 점수로 할 것이냐)
#25번을 바꿔본다.
#gridsearchCV 가 요구하는 형식이 있어 wrapper한후에 넣어준다.
#점수 - 민감도, 정밀도, 특이도, f1등

#greater_is_better=True 결과가 클수록 좋다.
#greater_is_better=false 결과가 작을수록 좋다.
#에러값은 작을 수록 좋다. 

grid_obj.fit(X_train_scaled, y_train)
scores = grid_obj.cv_results_['mean_test_score'].reshape(len(n_estimators), len(min_samples_leaf)) #사각형으로 리쉐입
#cv_results_ - 모든 결과담고 있다. 평균점수
plt.figure(figsize=(8,6))
plt.subplots_adjust(left=.2,right=0.95, bottom=0.15, top=0.95) #마진주기
plt.imshow(scores, interpolation='nearest',cmap=plt.cm.hot) #이미지로 출력, cmap제공되는 팔레트 사용
#interpolation='nearest' 보간법 가까운놈으로 출력해라 (반반쪼개진 사각형으로 출력이 되는것)
plt.xlabel('n_estimators')
plt.ylabel('min_samples_leaf')
plt.colorbar()
plt.xticks(np.arange(len(n_estimators)), n_estimators)
plt.yticks(np.arange(len(min_samples_leaf)), min_samples_leaf)
plt.title('parameter tuning AUC score')
plt.show()
#시커먼놈들이 점수가 낮고 밝은 놈들이 점수가 높다.
#n_estimators 는 작은놈들이 좋았다. (예측횟수)
```
