```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000, n_features=4, n_informative=2, random_state=0, shuffle=False)
#변수가 4개
#n_informative=2 1/0으로 값을 2개의 범주로 나누었다.
print(X)
print(y)
clf = AdaBoostClassifier(n_estimators=100, random_state=0) #adaptive - 분류에러에 모델이 적응해간다.
#에러에 가중치를 줘서 모델을 학습한 다음 모델을 모델에 더하는 방법
clf.fit(X,y)
print('변수 중요도',clf.feature_importances_) #트리학습의 중요한것 - 변수중요도, 시각화
#4개의 변수의 변수중요도 출력
print(clf.predict([[0,0,0,0]])) #데이터 한 쌍을 입력
print(clf.score(X,y))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.RandomState(1)
X = np.linspace(0,6,100)[:,np.newaxis] #newaxis 차원을 새로 생성한다.(일렬로 늘어진값을 하나의 열의 데이터로 바꾼다.)
y = np.sin(X).ravel() + np.sin(6*X).ravel() + rng.normal(0,0.1,X.shape[0])
print(rng)
print(X)
print(y)
regr1 = DecisionTreeRegressor(max_depth=4)
regr2 = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4), n_estimators=300, random_state=rng)
regr1.fit(X,y)
regr2.fit(X,y)
y_1 = regr1.predict(X)
y_2 = regr2.predict(X)
plt.figure()
plt.scatter(X,y, c='k', label='train data')
plt.plot(X,y_1, c='g',label='DTR', linewidth=2)
plt.plot(X,y_2, c='r',label='DTR300', linewidth=2)
plt.xlabel='data'
plt.ylabel='target'
plt.title('트리와 부스트의 차이')
plt.legend()
plt.show()

#처음에 직선으로 예측하다가. adaboost를 통해 회귀 분류 300개의 리그레서를 통해서 곡선으로 피팅시켰다.
#회귀 : 모델을 약한모델에서 강한모델로 만든다.
```

```python
from sklearn.ensemble import GradientBoostingRegressor
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston
from sklearn.metrics import mean_absolute_error
```

```python
boston = load_boston()
#bunch - .data, .target, .feature_names 에 정보 분리되어 저장
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.Series(boston.target)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size=0.7)
regressor = GradientBoostingRegressor(max_depth=2, n_estimators=4, learning_rate=1.0) #트리 : 깊이우선, 넓이우선
#GradientBoostingRegressor 넓이 우선방식으로 밑으로 내려가는 방식보다는 옆으로 넓게 깊이가 낮게 하는 방식으로 트리를 생성
#n_estimators 모델생성갯수, 
#learning_rate = 학습율이 나옴(정도), 학습율이 높다면 수렴하지 못함(최적화가 안된다.), 학습율이 작은경우 시간이 오래걸림
#learning_rate - hyper parameter이기때문에 입력만 해주면 된다.

regressor.fit(X_train, y_train)

#최대깊이, 최대 잎개수 들 설정가능
```

```python
regressor.loss_
#cost = loss , 현재 w,b값이 학습데이터를 얼마나 잘 표현하는가
#예측한 직선상의 y값과 실제 y점의 값의 오차로 본다.
```

```python
regressor.feature_importances_
#변수 13개에 대한 변수 중요도가 출력되었다.
#변수중요도를 이용해서 변수선택이 가능하다.
```

```python
regressor.train_score_ #error = 분산제곱차이다.(예측값과 실제값의 잔차의 제곱)
#scikits = model갯수 = estimators
#에러가 감소되는 것을 확인할 수 있다.
```

```python
#estimators를 높여도 에러가 많이 줄지 않을때
#early-stopping(XGBoost) 를실행(자동 가지치기 기능, 결측치를 자동으로 해결해주는 기능을 가졌다.)
```

```python
regressor.estimators_[0][0].predict(X_test)
#3개의 모델중 1번째의 모델
```

```python
errors = [mean_squared_error(y_test, y_pred) for y_pred in regressor.staged_predict(X_test)]
#staged_predict = estimator를 여러개를 stage라고 한다.
#모델갯수에 대해서 각기 테스트를 한것
#실 테스트데이서에 대해서 각 stage별로 예측을 하고 mean_squared_error를 낸것이다.
#각 스테이지 별로 테스트를 했다.
print(errors)
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.keys())
print(boston.data.shape)
print(boston.feature_names)
print(boston.DESCR)
```

```python
import pandas as pd
data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
data.head()
```

```python
data['price'] = boston.target
data.describe()
print(data.info())
```

```python
import xgboost as xgb
import numpy as np
#boosting 경사하강법으로 학습을 한다.(러닝레이트 존재)
#규제를 주기위해 alpha값 설정 - L1규제
#lambda - L2규제
from sklearn.metrics import mean_squared_error
X,y = data.iloc[:,:-1], data.iloc[:,-1]
print(X)
print(y) #price열의 데이터
data_dmatrix = xgb.DMatrix(data=X, label=y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
#XGBRegressor : scikits와 연결객체 -> gridsearchCV
#objective = 'reg:linear' 선형회귀 모델을 쓰겠다.
#분류 - objective = 'reg:tree'
#colsample_bytree 트리를 만들때 몇개의 특징으로 만들 것이냐 고려할 특징의 수
xgb_reg = xgb.XGBRegressor(objective = 'reg:linear', colsample_bytree=0.3, learning_rate=0.1, max_depth=5, alpha=10, n_estimators=10)
#값을 크게 주면 강한 규제를, 값을 작게 주면 약한 규제를 줄 수 있다.
xgb_reg.fit(X_train, y_train)
preds = xgb_reg.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, preds)) #RMSE(sqrt해줌)
print("RMSE : %f"%(rmse))
```

```python
import matplotlib.pyplot as plt
xgb.plot_tree(xgb_reg, num_trees=0)
plt.rcParams['figure.figsize']=[500,500]
plt.show()
```

```python
params = {'objective':'reg:linear','colsample_bytree':0.3, 'learning_rate':0.1, 'max_depth':5, 'alpha':10}
cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3, num_boost_round=50, early_stopping_rounds=10, metrics='rmse', 
                   as_pandas=True, seed=123)
#cv - nfold = 데이터는 접는 구간 3개로 나눠라(한번씩 다 테스트 해보겠다.)
#50회에 걸쳐하겠다.
#early stopping 조기종료 변화가 작으면 멈춰라
#metrics 척도를 정함
#as_pandas 데이터 프레임으로 리턴해라.

print(cv_results.head())
print((cv_results['test-rmse-mean']).tail(1))
```
